```python
from glob import glob
import os
import json
from tqdm import tqdm
```

```python
# txt label 파일을 json으로 변환
# YOLO txt format: class_id cx cy w h was_nonT
label_dirs = [
    '../../data/precise_BC_cell_scoring/her2/labels/',
    '../../data/precise_BC_cell_scoring/er_pr/labels/',
]

for label_dir in label_dirs:
    if not os.path.exists(label_dir):
        print(f"디렉토리 없음: {label_dir}")
        continue
    
    txt_files = glob(os.path.join(label_dir, '*.txt'))
    print(f"{label_dir}: {len(txt_files)}개 txt 파일 발견")
    
    for txt_path in tqdm(txt_files):
        labels = []
        with open(txt_path, 'r') as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue
                parts = line.split()
                labels.append({
                    "class_id": int(parts[0]),
                    "cx": float(parts[1]),
                    "cy": float(parts[2]),
                    "w": float(parts[3]),
                    "h": float(parts[4]),
                    "was_nonT": int(parts[5])
                })
        
        json_path = txt_path.replace('.txt', '.json')
        with open(json_path, 'w') as f:
            json.dump(labels, f)
        
        # 변환 완료 후 원본 txt 삭제
        os.remove(txt_path)
    
    print(f"{label_dir}: 변환 완료")

print("Done!")
```
